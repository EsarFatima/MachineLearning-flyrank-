## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
Classification (binary) — predicting is_declining_label — whose output probability doubles as a ranking/priority score. Not clustering (I already have a defined target, not an "what groups exist" question). Not pure ranking, because I don't have an observed rank, just a binary flag I can sort by.
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
Target: is_declining_label — but this is a defined proxy, not a truly observed outcome. It's derived from trend_direction, which is computed from trend_pct. Because the label is rule-built from those two columns, trend_pct and trend_direction can never be features — a model trained on them would just re-learn the rule, not the world. Since the starter CSV is a single 90-day snapshot with no forward-looking window, there's no genuinely observed "declined next quarter" label available here — is_declining_label is the best proxy I have in this dataset.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
Precision@K (top 200 items ranked by predicted decline probability), compared against the base decline rate. This matches how an editor actually uses the output — they work through the top of a queue, not the whole list, so overall accuracy (dominated by the easy majority class) isn't defensible here. ROC-AUC as a secondary, threshold-free check. "Good" = precision@K meaningfully above base rate — I'm using 1.5x base rate as a working bar.
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
Decline risk isn't behind one clean threshold — it shows up as several weak, correlated signals moving together (CTR, engagement rate, scroll depth, AI-traffic share, position). A single-feature rule (e.g., "flag if CTR is below median") captures almost none of it; combining the signals does measurably better. That gap — real signal, but spread thin across many columns — is exactly what makes this a job for a model instead of an if-statement.
```
