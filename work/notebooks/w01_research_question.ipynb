## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
Unit of analysis: a page (content_id), summarized over its trailing 90-day window — not a day, not a client.
Decision: which pages a content reviewer should look at first when they have limited review capacity (e.g., top 20–50 candidates).
Action: for a flagged page, the reviewer rewrites title/meta, improves intent match, or investigates why engagement is weak.
Cost of a wrong call: this is decision-support, not causal — so "wrong" mostly means wasted reviewer time (false positive: flagged a page that was actually fine) or a missed opportunity (false negative: a genuinely under-capturing page never got reviewed). Be explicit that you're not claiming a rewrite will definitely recover traffic — you're claiming the page deserves a look.
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# 1. How many pages even have enough volume to trust a CTR read?
viable = df[(df["impressions_90d"] >= 500) & (df["avg_position"] > 0) & (df["avg_position"] <= 20)]
print(f"Pages with enough impressions + top-20 position: {len(viable)} / {len(df)} ({len(viable)/len(df):.1%})")

# 2. Does CTR actually vary meaningfully by position tier? (justifies tier-adjustment)
print(df.groupby("position_tier")["ctr"].mean())

# 3. Of the "viable" pages, how many look like real opportunities (low CTR despite decent position)?
low_ctr_opportunity = viable[viable["ctr"] < 0.5]
print(f"Low-CTR opportunity candidates: {len(low_ctr_opportunity)} ({len(low_ctr_opportunity)/len(viable):.1%} of viable pages)")
```
